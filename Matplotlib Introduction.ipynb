# General Concepts

## Figure
A figure is a canvas which can contain multiple plots or axes. 

## Axes
This is generally what we think of as a plot. Each axes has a title, an x-label and a y-label

## Artist
Most artists are tied to axes. An artist is everything we can see on a figure, like `Text` objects, `Line2D` objects, and `collection` objects.

```python
from matplotlib import pyplot as plt
import numpy as np
```

# Basic Plots

```python
plt.plot([1, 2, 3, 4],[1, 4, 9, 16])
plt.title('A Basic Plot')
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

We can also specify the size of the figure using `figsize`, where the value is a tuple of the length of rows and columns.

```python
plt.figure(figsize=(15,5))
plt.plot([1, 2, 3, 4],[1, 4, 9, 16])
plt.title('A Basic Plot')
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

You can also specify the color and line type of the plot

```python
plt.plot([1, 2, 3, 4],[1, 4, 9, 16], "go")
plt.title('A Basic Plot')
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

```python
plt.plot([1, 2, 3, 4],[1, 4, 9, 16], "g^")
plt.title('A Basic Plot')
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

You can also plot multiple lines at once

```python
plt.plot([1, 2, 3, 4],[1, 4, 9, 16], "g^", [1, 2, 3, 4], [2, 3, 4, 5], "bo")
plt.title('A Basic Plot')
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

# Subplots

We use `subplot()` to add more than one plot to a figure. The `subplot()` method takes three arguments: `nrows`, `ncols`, and `index`. 

```python
plt.subplot(1, 2, 1)
plt.plot([1,2,3,4], [1,4,9,16], "g^")
plt.title("1st subplot")

plt.subplot(1, 2, 2)
plt.plot([1,2,3,4], [2,3,4,5], "bo")
plt.title("2nd subplot")

plt.suptitle("My Subplots")
plt.show()
```

```python
# Let's try a different layout

plt.subplot(2, 1, 1)
plt.plot([1,2,3,4], [1,4,9,16], "g^")
plt.title("1st subplot")

plt.subplot(2, 1, 2)
plt.plot([1,2,3,4], [2,3,4,5], "bo")
plt.title("2nd subplot")

plt.suptitle("My Subplots")
plt.show()
```

However, the above method can be a little tedious for when we have many subplots. An alternative is to use the `subplots` method, which creates two objects: `figure` and `axes`. 

```python
x = np.arange(1,5)
y1 = x**2
y2 = x**3

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(6,6))
axes[0,1].plot(x, y1, 'go')
axes[1,0].plot(x, y2, 'r^')
axes[0,1].set_title("Squares")
axes[1,0].set_title("Cubes")
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(6,6))
i = 0
for ax in axes.ravel():
    ax.plot(x, x**i)
    ax.set_title('X^' + str(i))
    i += 1
plt.show()
```

# Other Types of Plots

## Bar Graphs

Instead of `plot()`, we use `bar()`

```python
categories = ["Cat A", "Cat B", "Cat C"]
x = range(len(categories)) # We have to convert the categories into an array of integers, because it can't take strings
category_values = [80, 45, 110]

plt.bar(x, category_values, color='green')
plt.title("A Simple Bar Graph")
plt.xlabel("Categories")
plt.xticks(x, categories) # But here we can show the string value instead of the index
plt.ylabel("Values")
plt.show()
```

```python
# A horizontal Bar Graph with variance
categories = ["Cat A", "Cat B", "Cat C"]
x = range(len(categories))
category_values = [80, 45, 110]
variance = [10, 3, 24]

plt.barh(x, category_values, color='green', xerr=variance)
plt.title("A Simple Bar Graph")
plt.ylabel("Categories")
plt.yticks(x, categories)
plt.xlabel("Values")
plt.show()
```

## Pie Charts

We use `pie()` to make a pie chart

```python
firms = ["Firm A", "Firm B", "Firm C", "Firm D"]
market_share = [20, 10, 35, 35]
plt.pie(market_share, labels=firms)
plt.axis('equal')
plt.show()
```

```python
# We can also emphasize a slice with Explode

firms = ["Firm A", "Firm B", "Firm C", "Firm D"]
market_share = [20, 10, 35, 35]
explode = [0, 0, 0, 0.1]

plt.pie(market_share, labels=firms, explode=explode)
plt.axis('equal')
plt.show()
```

## Histogram

We can plot a histogram of a variable using `hist()`

```python
x = np.random.randn(1000)

plt.title("Histogram")
plt.xlabel("Random Data")
plt.ylabel("Frequency")
plt.hist(x)
plt.show()
```

## Scatter Plot

```python
height = np.array([167, 170, 149, 165, 155, 180, 166, 149, 159])
weight = np.array([86, 74, 66, 78, 89, 80, 71, 68, 81])
plt.xlim(140, 200) # set the bounds of the x-axis
plt.ylim(60, 100)
plt.scatter(height, weight)
plt.title("Scatter Plot")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

# Using Seaborn Color Schemes

https://seaborn.pydata.org/tutorial/color_palettes.html

```python
import seaborn as sns

sns.set() # Use the default color palette
plt.plot([1, 2, 3, 4],[1, 4, 9, 16])
plt.title('A Basic Plot')
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

```python
sns.set_palette("Set2") # Set a new default palette

firms = ["Firm A", "Firm B", "Firm C", "Firm D"]
market_share = [20, 10, 35, 35]
explode = [0, 0, 0, 0.1]

plt.pie(market_share, labels=firms, explode=explode)
plt.axis('equal')
plt.show()
```

```python
sns.set_palette(["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"])

firms = ["Firm A", "Firm B", "Firm C", "Firm D"]
market_share = [20, 10, 35, 35]
explode = [0, 0, 0, 0.1]

plt.pie(market_share, labels=firms, explode=explode)
plt.axis('equal')
plt.show()
```

```python
sns.set_palette((sns.cubehelix_palette(8)))
firms = ["Firm A", "Firm B", "Firm C", "Firm D"]
market_share = [20, 10, 35, 35]
explode = [0, 0, 0, 0.1]

plt.pie(market_share, labels=firms, explode=explode)
plt.axis('equal')
plt.show()
```
