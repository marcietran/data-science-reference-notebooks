Why use Seaborn? It provides a simple way to create beautiful graphs in Python. Seaborn is built on top of `matplotlib`, and it is easy to use both of them together. Seaborn can make your existing `matplotlib` plots more attractive by simply using its color schemes, and it also extends `matplotlib`'s capabilities with new types of plots

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

[Tutorial Source](https://www.datacamp.com/community/tutorials/seaborn-python-tutorial)

[Official Seaborn Documentation](https://seaborn.pydata.org/api.html)

```python
# Load the Iris Dataset
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
# Construct the Plot
sns.swarmplot(x="species", y="petal_length", data=iris)

# Show it
# (Even though we're using seaborn, matplotlib still underlies it, so we use plt.show())
plt.show()
```

```python
# Let's look at Titanic Data
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.factorplot("class", "survived", "sex", data=titanic, kind="bar", palette="muted", legend=True)
plt.show()
```

```python
# Let's experiment with "Context"
sns.set() # Reset to default parameters
# The other options for context are "paper", "notebook", "talk", and "poster", in ascending size
sns.set_context("poster")
sns.swarmplot(x="species", y="petal_length", data=iris)
plt.show()
```

```python
# You can also use set_context for things like font_scale and axes labelsize

sns.set()
sns.set_context("notebook", font_scale=1.5, rc={"axes.labelsize": 15})
sns.swarmplot(x="species", y="petal_length", data=iris)
plt.show()
```

```python
tips = sns.load_dataset("tips")

with sns.axes_style("white"): # you can use "with" to temporarily set a plot style
    plt.subplot(211)
    sns.swarmplot(x="species", y="petal_length", data=iris)
    
plt.subplot(212)
sns.violinplot(x="total_bill", data=tips)
plt.show()
```

There are two types of functions for seaborn: **axes-level** and **figure-level**. 

**axes-level** functions are `regplot()`, `boxplot()`, `kdeplot()`, etc. These take an explict `ax` argument and return an `Axes` object. 

**figure-level** functions are `lmplot()`, `factorplot()`, and `jointplot()`. These can potentially include `Axes` and organized in a "meaningful" way

```python
#axes-level
sns.boxplot(x="total_bill", data=tips)
```

```python
# Setting Figure Size for Axes Level Functions
fig, ax = plt.subplots(figsize=(6,6))
sns.swarmplot(x="species", y="petal_length", data=iris, ax=ax)
plt.show()
```

```python
# For figure-level functions, you can set figure size with "size" and "aspect"
# aspect controls width
# size controls overall scale
sns.factorplot("class", "survived", "sex", data=titanic, kind="bar", size=6, aspect=2, palette="muted", legend=False)
plt.show()
```
